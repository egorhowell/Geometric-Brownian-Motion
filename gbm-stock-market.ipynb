```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns

drift = 0.003
vola = 0.075
start_price = 100
N = 2000000
var_vola = []
var_drift = []
price=[]
x = []
p= []
np.random.seed(112)


for i in range(N):
    var_vola.append(vola * np.random.normal(0, 1))
    var_drift.append(drift * np.random.normal(0, 1))
    
    
for i in range(N):
    change = var_drift[i]*start_price + var_vola[i]*start_price*np.random.normal(0, 1)
    new = start_price + change*0.01
    start_price = new
    x.append(i)
    price.append(start_price)

price.sort()
for i in range(N):
    p.append(i/N)

plt.plot(price,p)
```
